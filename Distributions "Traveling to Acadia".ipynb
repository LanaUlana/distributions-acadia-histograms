TASK:

As employees at a travel agency we want to inform our customers of the best time to visit one of our favorite vacation locations, Acadia, Maine. In this project, we will use flower bloom data, and flight information to recommend the best time of year for someone to make a trip to Maine.

In script.py, we’ve imported the following data:

* in_bloom is a NumPy array containing 900 numbers. Each number is a day of the year, which represents the average start date of a flower blooming.
* flights is a NumPy array with 11,000 numbers. Each number is a day of the year, which represents a flight from our hometown to airports near Acadia, Maine.

Our goal is to create two histograms, each displaying the frequency of an occurrence each day of the year (either flights or flower blooms).

We will use the in_bloom variable to find a count of the number of flowers that start blooming each day of the year.

We will use the flights variable to find a count of the number of flights that occur each day of the year.


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# loading in data
in_bloom = pd.read_csv("/home/ulyana/Documents/distributions_acadia/in-bloom.csv")
flights = pd.read_csv("/home/ulyana/Documents/distributions_acadia/flights.csv")

# Ploting the histograms
plt.figure(1)
plt.subplot(211)
bins = np.linspace(0, 365, 10)

plt.hist(flights, range=(0, 365), bins=365,  edgecolor="black")
plt.title("Flights")
plt.xlabel("Days of the year")
plt.ylabel("Number of flights")

plt.subplot(212)
plt.hist(in_bloom, range=(0, 365), bins=365,  edgecolor="black")
plt.title("Blooming flowers")
plt.xlabel("Days of the year")
plt.ylabel("Number of flowers")

plt.tight_layout()
plt.show()
```

Based on these histograms, we would propose to go between days 250 and 300, which is the Fall. The flight’s histogram shows fewer flights between these days, so there wouldn’t be many people, while the flowers histogram indicates there should be a number of flowers in bloom during this time.
